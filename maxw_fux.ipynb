```python
#Little script that re-create Maxwellian and Maxwellian flux distributions
import numpy as np
import matplotlib.pyplot as plt



Np = 500000;  #Number of particle to create
Te = 20;     #[eV] Electron distribution temperature
me = 9.109e-31; #[kg] electron mass
q = 1.6021765650e-19; #[C] electron charge
kb = 1.3806488e-23;  #Blozman constant
eps_0 = 8.8548782e-12; #Vaccum permitivitty

tabe = np.zeros(Np)
```

```python
def velocity_maxw_flux():
    import random
    v_Te = np.sqrt(2*q*Te/me)

    R = random.random()
    v = (v_Te*np.sqrt(-np.log(R)))
    return v

def velocity_maxw():
    import random
    v_Te = np.sqrt(q*Te/me)

    W = 2
 
    
    while (W >= 1 or W <= 0):
        R1 = (random.random()*2 -1 )
        R2 = (random.random()*2 -1 )
        
        W = R1**2 + R2**2
    W = np.sqrt( (-2*np.log(W))/W)
        
    v = v_Te*R1*W
    return v
```

```python
Np = 10000000;  #Number of particle to create
Te = 20;     #[eV] Electron distribution temperature
tabe_v = np.zeros(Np)

#init velocitises
for i in np.arange(Np):
    tabe_v[i] = velocity_maxw_flux()

#calculate positions
dt =  10
tabe_x = np.zeros(Np)

import random
for i in np.arange(Np):
    R = random.random()
    R *= dt
    tabe_x[i] = R*tabe_v[i]
```

```python
plt.hist(tabe_v, 100, normed=1, facecolor='green', alpha=0.75)
plt.title("Probability fonction of velocity")
plt.ylabel("probabitlity fonction")
plt.xlabel("velocity v [m/s]")
plt.show()

v_mean = sum(tabe_v)/Np
Te_distrib = me/2*np.sum((tabe_v-v_mean)**2)/Np

print("the mean velocity is {} m.s-1".format(v_mean))
#print("the temperature of the distribution is {} eV".format(2*Te_distrib/q))
    

plt.scatter(tabe_x[1:int(Np/10)],tabe_v[1:int(Np/10)])
plt.title("Phase space distribution after dt = {}".format(dt))
plt.xlabel("position x [m]")
plt.ylabel("velocity v [m/s]")
plt.show()

#local velocity distribution
dx = 100000
plt.hist(tabe_v[tabe_x <dx], 50, normed=1, facecolor='green', alpha=0.75)
plt.title("distribution function averaged between 0 and {} m".format(dx))
plt.ylabel("distribution fonction")
plt.xlabel("velocity v [m/s]")
plt.show()
```
