```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Siz = pd.read_csv("Siz.txt",header = None)
```

```python
plt.plot(Siz)
plt.show()
```

```python
import numpy as np

Np = len(Siz)
X = [i/Np for i in np.arange(Np)]
```

```python
z = np.polyfit(X, Siz, 20)
p = np.poly1d([i[0] for i in z])
```

```python
plt.plot(p(X))
plt.plot(Siz)
plt.show()
```

```python
from scipy.optimize import curve_fit

def expo(x,a,b,c,d):
    
    return np.array((a*x+1)*b*np.exp((-(x-c)**2)/d))

param_bounds=([-np.inf,-np.inf,0,0],[np.inf,np.inf,1,np.inf])

popt, pcov = curve_fit(expo, np.array(X).ravel(), np.array(Siz).ravel(), p0=(0,1e20,0.1,0.1),bounds=param_bounds)
    
```

```python
popt=[-0,1.37e24,0.56,0.015]
plt.plot(X,[expo(i,popt[0],popt[1],popt[2],popt[3]) for i in X])
print(popt)
plt.plot(X,Siz)
plt.show()
plt.savefig("")
```
