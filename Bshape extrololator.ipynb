```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Nx, Ny = 255, 100
taby = np.arange(Ny)
Nxin, Nyin = 12,12
tabyin = np.arange(Nyin)

Bsheap = np.ones((Nx,Ny))
Binput = np.ones((Nxin,Nyin))
```

```python
for j in np.arange(Nxin):
    for i in tabyin:

        Binput[j,i] = np.exp(-(i-Nyin/2)**2/(10*Nyin)) * np.exp(-(j-Nxin/2)**2/(10*Nxin))
    
Binput /= Binput.max()

for j in np.arange(Nx):
    for i in taby:
        x = (i*(Nyin-2)/Ny)+1
        y = (j*(Nxin-2)/Nx)+1
        
        y1 = int(y)
        y2 = y1 +1
        
        x1 = int(x)
        x2 = x1+1

        Bsheap[j,i] = ((x2-x)*(y2 - y)*Binput[y1,x1] + (x - x1)*(y2 - y) *Binput[y1,x2] +
                       (x2-x)*(y - y1)*Binput[y2,x1] + (x - x1)*(y - y1) *Binput[y2,x2] )
```

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(Binput[:,1:],interpolation="gaussian",aspect=1)
#plt.imshow(Binput)

ax = plt.subplot(223)
plt.imshow(Bsheap[:,:],interpolation="nearest",aspect=0.4)


plt.subplot(222)

plt.plot(taby/Ny,Bsheap[1,:],'+')


plt.plot((tabyin[1:]-1)/(Nyin-2),Binput[1,1:],'o')
```
